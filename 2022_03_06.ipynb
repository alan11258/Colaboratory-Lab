```python
import requests;

url = "https://opendata.epa.gov.tw/api/v1/PM25?%24skip=0&%24top=1000&%24format=json";

res = requests.get( url );
resCont = res.content;
resJson = res.json();

# city = "金門";
city = input("請輸入想查詢的區域:");

for i in resJson:
  # print( i );
  site_name = i['SiteName'];
  monitor_Date = i['MonitorDate'];
  concentration = i['Concentration'];
  itemUnit = i['ItemUnit'];
  if site_name == city and int( concentration ) > 10:
    report = f"地點:{ site_name }, 監測日期:{ monitor_Date }, PM2.5濃度:{ concentration }, 單位:{ itemUnit }";
    print( report );

```

```python

# <div class="hidden-phone print_show" style="text-indent:30px;">
#  美金 (USD)
# </div>

# 安裝指定模組
# !pip install pyquery;

from pyquery import PyQuery;

web_url = "https://rate.bot.com.tw/xrt?Lang=zh-TW";

# class="hidden-phone print_show"

name_list = html("div.hidden-phone.print_show").text().split();

zh_name_list = [];

for idx, name in enumerate( name_list ):
  if idx % 2 == 0:
    # print( name );
    zh_name_list.append( name );

# print( zh_name_list );

# <td data-table="本行現金買入" class="rate-content-cash text-right print_hide">27.715</td>

# buy_list = html("td.rate-content-cash.text-right.print_hide").text();
# buy_list = html("td[data-table='本行現金買入']").text().split()

buy_list = html("td[data-table='本行現金買入'].print_hide").text().split();
sell_list = html("td[data-table='本行現金賣出'].print_hide").text().split();

# print( buy_list );
# print( sell_list );
table = {};

for name, buy, sell in zip( zh_name_list, buy_list, sell_list):
  # print( name, buy, sell );
  table[name] = {
      "buy" : buy,
      "sell" : sell
  }

# print( table );

# search = "美金";

def fun_searchCurrenct():
  while True:
    search = input("請輸入查詢幣別:")

    if search not in zh_name_list:
      print( f"查無此幣別資訊:{search}" );
    else:
      buy = table[search]['buy'];
      sell = table[search]['sell'];

      result = f"{search}, 買入:{buy}, 賣出:{sell}";
      print( result );
      break;

fun_searchCurrenct();



```
